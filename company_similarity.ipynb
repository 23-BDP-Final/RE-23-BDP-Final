```python
import pandas as pd

# 검색한 company name과 dictionary의 key(company)의 단어 존재 여부에 따라 ExclusiveOr
def similarity(L1,L2):
  count=0
  for i in range(len(L1)):
    count+=int(not(L1[i]^L2[i]))
  return count

back = pd.read_csv('https://raw.githubusercontent.com/23-BDP-Final/RE-23-BDP-Final/main/data/content_back.csv')
back_word = pd.read_csv('https://raw.githubusercontent.com/23-BDP-Final/RE-23-BDP-Final/main/mapreduce/wordcount_back.txt',header=None)

back_word[['word', 'count']] = back_word[0].str.split('\t', n=1, expand=True)
back_word=back_word[['word', 'count']]

# wordcount가 많은 상위 10개 단어 추출
n=10

# 해당 단어가 있는지 없는지에 대한 column 생성 (현재는 0으로 채워져있다.)
for i in range(n):
  back['is_'+back_word['word'][i]]=0

# 데이터에 존재하는 영단어를 소문자로 변경 후, column에 T/F로 채운다.
for i in range(len(back)):
  for j in range(n):
    back['is_'+back_word['word'][j]][i]=(back_word['word'][j] in back['description'][i].lower())|(back_word['word'][j] in back['requirement'][i].lower())|(back_word['word'][j] in back['preferred'][i].lower())
word_list = []
for i in range(n):
  word_list.append('is_'+back_word['word'][i])

# key를 'company_name'으로, value는 단어 존재 여부들(T/F)
d=dict()
for j in range(len(back)):
  d[back['company_name'][j]]=back[word_list].iloc[j].tolist()

# 검색할 company_name 입력
target_company=input()

# 검색한 company name과 dictionary의 key(company)의 유사도를 계산
sim_list=[]
for i in d:
  sim_list.append((i,similarity(d[target_company],d[i])))

# 유사도를 기반으로 sort
sim_list.sort(key=lambda x:x[1],reverse=True)
sim_list.remove((target_company,n))
print(sim_list)
```
