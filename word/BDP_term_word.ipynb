```python
import pandas as pd
contents=pd.read_csv('https://raw.githubusercontent.com/23-BDP-Final/RE-23-BDP-Final/main/preprocessing/contents.csv')
contents
```

```python
#전체 description data를 하나의 1차원 리스트로 합치기
#전체 requirement data를 하나의 1차원 리스트로 합치기

import numpy as np
x = contents['description'].tolist()
y = contents['requirement'].tolist()
p = contents['preferred'].tolist()

contents_description = ''
for i in x:
  contents_description += i
contents_requirement = ''
for i in y:
  contents_requirement += i
contents_preferred = ''
for i in p:
  contents_preferred += i
```

```python
# 전체 data에서 wordcloud

# 전체 - description data
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import konlpy
from konlpy.tag import Okt
from collections import Counter
from PIL import Image

#불용어 제거 함수
t = Okt()
def twitter_tokenizer(text):
    return t.morphs(text)
def twitter_tokenizer_filter(text):
    mailst = t.pos(text,norm=True, stem=True)
    word_types_to_filter = {'Josa', 'Eomi', 'Punctuation', 'KoreanParticle'}
    r = [word[0] for word in mailst if word[1] not in word_types_to_filter]
    return ' '.join(r)

# 불용어 제거 / 한 글자인 단어 제거
tokens = t.nouns(twitter_tokenizer_filter(contents_description))
tokens = [word for word in tokens if len(word) > 1]
count = Counter(tokens)
tags = count.most_common(10)
print(type(tags))

dtags = dict(tags)
print(dtags)

# 단어 수 450개 이상인 단어만 필터링
dict_tags = {key: value for key, value in dtags.items() if value >= 450}

heart_mask = np.array(Image.open('/content/drive/MyDrive/python23/termpro/heart-01.png'))

wordcloud = WordCloud(font_path=fontpath, mask=heart_mask,background_color='white', max_font_size=40).generate(contents_description)

#Display the word cloud using matplotlib
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()


# 전체 - requirement data

# 불용어 제거 / 한 글자인 단어 제거
tokens = t.nouns(twitter_tokenizer_filter(contents_requirement))
tokens = [word for word in tokens if len(word) > 1]
count = Counter(tokens)
tags = count.most_common(10)
print(type(tags))

dtags = dict(tags)
print(dtags)

# 단어 수 450개 이상인 단어만 필터링
dict_tags = {key: value for key, value in dtags.items() if value >= 450}

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()


# 전체 - preferred data

# 불용어 제거 / 한 글자인 단어 제거
tokens = t.nouns(twitter_tokenizer_filter(contents_preferred))
tokens = [word for word in tokens if len(word) > 1]
count = Counter(tokens)
tags = count.most_common(10)
print(type(tags))

dtags = dict(tags)
print(dtags)

# 단어 수 450개 이상인 단어만 필터링
dict_tags = {key: value for key, value in dtags.items() if value >= 450}

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# 구글 번역기 api 사용
from googletrans import Translator

translator = Translator()

# contents_description 한글 -> 영어로 번역
contents_description = []
for i in x:
  result1 = translator.translate(i, src='ko', dest='en')
  contents_description.append(result1.text)
```

```python
# contents_requirement 한글 -> 영어로 번역
contents_requirement = []
for i in y:
  result2 = translator.translate(i, src='ko', dest='en')
  contents_requirement.append(result2.text)
```

```python
# contents_preferred 한글 -> 영어로 번역
contents_preferred = []
for i in p:
  result = translator.translate(i, src='ko', dest='en')
  contents_preferred.append(result.text)
```

```python
# contents_description txt 파일로
with open('contents_description.txt','w',encoding='UTF-8') as f:
    for i in contents_description:
        f.write(str(i))
```

```python
# contents_requirement txt 파일로
with open('contents_requirement.txt','w',encoding='UTF-8') as f:
    for i in contents_requirement:
        f.write(str(i))
```

```python
# contents_preferred txt 파일로
with open('contents_preferred.txt','w',encoding='UTF-8') as f:
    for i in contents_preferred:
        f.write(str(i))
```
