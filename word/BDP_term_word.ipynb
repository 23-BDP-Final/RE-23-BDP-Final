```python
import pandas as pd
contents=pd.read_csv('/content/drive/MyDrive/python23/termpro/contents.csv')
contents
```

```python
#전체 description data를 하나의 1차원 리스트로 합치기
#전체 requirement data를 하나의 1차원 리스트로 합치기

import numpy as np
x = contents['description'].tolist()
y = contents['requirement'].tolist()
contents_description = ''
# contents_description=np.ravel(contents_description, order='K')
for i in x:
  contents_description += i

contents_requirement = ''
for i in y:
  contents_requirement += i
```

```python
with open('contents_description.txt','w') as f:
  f.write(contents_description)

with open('contents_requirement.txt','w') as f:
  f.write(contents_requirement)
```

```python
# 전체 data에서 wordcloud

# 전체 - description data
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import konlpy
from konlpy.tag import Okt
from collections import Counter
from PIL import Image

#불용어 제거 함수
t = Okt()
def twitter_tokenizer(text):
    return t.morphs(text)
def twitter_tokenizer_filter(text):
    mailst = t.pos(text,norm=True, stem=True)
    word_types_to_filter = {'Josa', 'Eomi', 'Punctuation', 'KoreanParticle'}
    r = [word[0] for word in mailst if word[1] not in word_types_to_filter]
    return ' '.join(r)

# for w in word_tokens:
#     if w not in stop_words:
#         result.append(w)

# 불용어 제거 / 한 글자인 단어 제거
tokens = t.nouns(twitter_tokenizer_filter(contents_description))
tokens = [word for word in tokens if len(word) > 1]
print(tokens)
count = Counter(tokens)
tags = count.most_common(10)
print(type(tags))

dtags = dict(tags)
print(dtags)

# 단어 수 450개 이상인 단어만 필터링
dict_tags = {key: value for key, value in dtags.items() if value >= 450}

heart_mask = np.array(Image.open('/content/drive/MyDrive/python23/termpro/heart-01.png'))

wordcloud = WordCloud(font_path=fontpath, mask=heart_mask,background_color='white', max_font_size=40).generate(contents_description)

#Display the word cloud using matplotlib
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# 전체 - requirement data

# 불용어 제거 / 한 글자인 단어 제거
tokens = t.nouns(twitter_tokenizer_filter(contents_requirement))
tokens = [word for word in tokens if len(word) > 1]
count = Counter(tokens)
tags = count.most_common(10)
print(type(tags))

dtags = dict(tags)
# print(dtags)

# 단어 수 450개 이상인 단어만 필터링
dict_tags = {key: value for key, value in dtags.items() if value >= 450}

plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
